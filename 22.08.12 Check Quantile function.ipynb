```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfinv
import scipy.stats as scs
```

```python
Quantile_Lorentzian = lambda x,d,N : np.array([d*np.tan(np.pi/2 * (2*i - N - 1)/(N+1)) + x for i in range(1,N+1)])
Quantile_Normal = lambda mu,sigma,N : np.array([mu +sigma*(2**0.5)*erfinv((2*i - N - 1)/(N+1)) for i in range(1,N+1)])
```

```python
x = np.arange(0,100,0.1)
plt.plot(x,erfinv(x))
# erf(10)
```

```python
plt.hist(Quantile_Normal(0,1,50000),bins=100);
plt.hist(np.random.normal(0,1,50000),bins=100);
```

```python
Q_Lorentzian = Quantile_Lorentzian(0,1,500)
P_Q = Q_Lorentzian[np.where(Q_Lorentzian>0)]
M_Q = Q_Lorentzian[np.where(Q_Lorentzian<0)]
np.random.seed(0)
R_Lorentzian = scs.cauchy.rvs(0,1,500)

P_R = R_Lorentzian[np.where(R_Lorentzian>0)]
M_R = R_Lorentzian[np.where(R_Lorentzian<0)]
```

```python
plt.hist([Q_Lorentzian[abs(Q_Lorentzian)<5],R_Lorentzian[abs(R_Lorentzian)<5]],bins=100, alpha=0.7, histtype='stepfilled',
         label = ['Quantile function','Random sampling']);
plt.legend()
```
